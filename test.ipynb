```python
from business_rules_genai.actions import BaseActions
from business_rules_genai.fields import FIELD_NUMERIC, FIELD_TEXT
from business_rules_genai.operators import NumericType, StringType, BooleanType

class CompanyActions(BaseActions):

    def __init__(self, company):
        self.company = company

    # DO NOT DELETE THESE FUNCTIONS
    def set_value_numeric(self, value):
        return NumericType(value)
    
    def set_value_string(self, value):
        return StringType(value)

    def always_true(self):
        return BooleanType(True)
    
    def add(self, value1, value2):
        if isinstance(value1, NumericType):
            value1 = value1.value
        if isinstance(value2, NumericType):
            value2 = value2.value
        return NumericType(value1 - value2)
    
    def minus(self, value1, value2):
        if isinstance(value1, NumericType):
            value1 = value1.value
        if isinstance(value2, NumericType):
            value2 = value2.value
        return NumericType(value1 - value2)
    
    def mult(self, value1, value2):
        if isinstance(value1, NumericType):
            value1 = value1.value
        if isinstance(value2, NumericType):
            value2 = value2.value
        return NumericType(value1 * value2)
    
    def divide(self, value1, value2):
        if isinstance(value1, NumericType):
            value1 = value1.value
        if isinstance(value2, NumericType):
            value2 = value2.value
        if value2 == 0:
            return NumericType(0)
        return NumericType(value1 / value2)
    
    # ADD NEW FUNCTIONS HERE
    # Functions must return NumericType or StringType or BooleanType
    def percentage(self, value1, value2):
        if isinstance(value1, NumericType):
            value1 = value1.value
        if isinstance(value2, NumericType):
            value2 = value2.value
        if value2 == 0:
            return NumericType(0)
        return NumericType((value1 / value2) * 100)
```
